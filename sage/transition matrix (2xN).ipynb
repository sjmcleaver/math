```python

def B(n):
    R = PolynomialRing(ZZ, n + 1, "z")
    blocks = []
    for i in range(n):
        blocks.append(Matrix(R, [
            [R.gen(0)^2, 0, 0],
            [R.gen(0)^2, R.gen(0) * R.gen(i + 1), 0],
            [R.gen(0)^2, 0, R.gen(0) * R.gen(i + 1)]
        ]))
    return block_diagonal_matrix(blocks)

def E(n):
    R = PolynomialRing(ZZ, n + 1, "z")
    rows = []
    for i in range(3 * n):
        if i % 3 == 0:
            row = []
            for j in range(n):
                if j == (i / 3) + 1:
                    row += [0, R.gen(0) * R.gen((i / 3) + 1), R.gen(0) * R.gen((i / 3) + 1)]
                else:
                    row += [0, 0, 0]
            rows.append(row)
        else:
            rows.append([0] * 3 * n)
    return Matrix(R, rows)

def T(n):
    return B(n) + E(n)

[(B(4) * E(4))^n for n in range(4)]

# T^n = (B + E)^n
# E^2 = 0

# M_0 := B, M_1 :=  E
# T^n = Sum_(length n bit strings s) Prod_i M_(s_i)
#     = Sum_(length n bit strings s not containing 11) Prod_i M_(s_i)
# 
# { (0 -> 0), (0 -> 1), (1 -> 0) }
# [ B, E ]
# [ B, 0 ]
```
